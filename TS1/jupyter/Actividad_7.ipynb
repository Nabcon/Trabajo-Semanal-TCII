Nuestra función transferencia nos queda de la siguiente manera:

```python
from sympy import symbols, init_printing, Eq

from IPython.display import display

R_21 = symbols('R_{1}/R_2')
s = symbols("S")
C = symbols("C")

R_1, R_2, R_3 = symbols('R_1 R_2 R_3')
R_3C = symbols('1/(R_{3}C)')
T_s = symbols('T({S})')
Wo = symbols('\omega_0')



display(Eq(T_s,((s) -( R_21*R_3C) ) / (s + R_3C)))
display(Eq(T_s,((s) -( R_21*Wo) ) / (s + Wo)))
```

Donde *normalizando en frecuencia* nos queda la siguiente expresión:

```python
display(Eq(T_s,((s) -( R_21) ) / (s + 1)))
```

### Punto N°3

A continuación vamos a poder ver la función transferencia normalizada.

Podemos notar que el valor de la transferencia depende de la constante $k$ donde:

\begin{equation}
k = \frac{R_2}{R_1}
\end{equation}

Para distintos valores de $k$ obtenemos diferentes gráficos, siendo para $k=1$ donde obtenemos un **filtro pasa todo** de ganancia unitaria, visualizado en color naranja

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import scipy.signal as sig
from splane import pretty_print_lti, analyze_sys

#Presentacion de los graficos
mpl.rcParams['figure.figsize'] = (10,10)
fig_font_size = 16
plt.rcParams.update({'font.size':fig_font_size})


all_sos = []
#Le doy distintos valores a k para ver como se comportan los graficos
valores_k = [0.5 , 1 , 1.5]


for k in valores_k:
        
    num = np.array([1,-k])
    den = np.array([1,1])
    
    my_sos = sig.TransferFunction(num,den)
    
    pretty_print_lti(my_sos)

    all_sos += [my_sos]
    
plt.close('all')
analyze_sys(all_sos, sys_name=valores_k)
```
