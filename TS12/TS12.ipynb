### Ejercicio 1

```python
import sympy as sp
import splane as tc2
from schemdraw import Drawing
from schemdraw.elements import  Capacitor, Inductor

# YY

# Resolución simbólica

s = sp.symbols('s ', complex=True)

# Sea la siguiente función de excitación
YY = (3*s**3+7*s)/((s**2+2)*(s**2+5))

w = 1

Z2, ZC1 = tc2.remover_polo_dc(1/YY, omega_zero = w)

C1 = 1/(ZC1*s)

Y4, Zt1, L1, C2 = tc2.remover_polo_jw(1/Z2, omega = w, isImpedance = False)

Z6, ZL2 = tc2.remover_polo_infinito(1/Y4)

L2 = ZL2/s

Z8, ZC3 = tc2.remover_polo_dc(Z6)

C3 = 1/(ZC3*s)

# Dibujamos la red resultante:
    
d = Drawing(unit=3)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                               port_name = 'In', 
                               voltage_lbl = ('+', '$V_1$', '-'), 
                               current_lbl = '$I_1$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Y',  
                                          YY, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)

d = tc2.dibujar_elemento_serie(d, Capacitor, C1)

d = tc2.dibujar_tanque_derivacion(d, L1, C2)

d = tc2.dibujar_elemento_serie(d, Inductor, L2)

d = tc2.dibujar_elemento_serie(d, Capacitor, C3)

d = tc2.dibujar_puerto_salida(d, 
                              port_name = 'Out', 
                              current_lbl = '$I_2$' )

display(d)
```

### Ejercicio 2 - Topologia Z11

Verificaremos si los valores y componentes obtenidos son correctos

```python
from schemdraw.elements import  Resistor

# ZZ

s = sp.symbols('s ', complex=True)

# Sea la siguiente función de excitación

ZZ = ((s+2)*(s+4))/((s+3)*(s+1))

# Halle los valores de los componentes de la topología resultante.

# remoción parcial para que el siguiente tanque R1-C1 resuenen a 6 r/s

# Consignas del ejercicio: resonancias de dos tanques RC
sigma1 = 1

# La topología circuital guía las remociones:

Z2, k1, R1, C1 = tc2.remover_polo_sigma(ZZ, sigma1, isImpedance = True)

Z4, R2 = tc2.remover_valor_en_infinito(Z2)

Y6, Yc2 = tc2.remover_polo_infinito(1/Z4)

C2 = Yc2/s

R3 = 1/Y6

# Dibujamos la red resultante:
    
d = Drawing(unit=3)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                               port_name = 'In', 
                               voltage_lbl = ('+', '$V_1$', '-'), 
                               current_lbl = '$I_1$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Z_{11}',  
                                          ZZ, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)

d = tc2.dibujar_tanque_RC_serie(d, R1, C1)

d = tc2.dibujar_elemento_serie(d, Resistor, R2)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C2)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, R3)

d = tc2.dibujar_puerto_salida(d, 
                              port_name = 'Out', 
                              current_lbl = '$I_2$' )

display(d)
```

### Ejercicio 2 - Topologia Y22

Verificaremos si los valores y componentes obtenidos son correctos

```python
import sympy as sp
import splane as tc2
from schemdraw import Drawing
from schemdraw.elements import  Resistor

# Resolución simbólica

s = sp.symbols('s ', complex=True)

# Sea la siguiente función de excitación
YY =  ((s+2)*(s+4))/(s+3)

# Halle los valores de los componentes de la topología resultante.

# remoción parcial para que el siguiente tanque R1-C1 resuenen a 6 r/s

# Consignas del ejercicio: resonancias de dos tanques RC
sigma1 = 1

# La topología circuital guía las remociones:
    
Y2, Yc1 = tc2.remover_polo_infinito(YY)

C1 = Yc1/s

Y4, G1 = tc2.remover_valor(Y2, sigma_zero = sigma1)

R1 = 1/G1

Z6, k1, R2, C2 = tc2.remover_polo_sigma(1/Y4, sigma1, isImpedance = True)   

R3 = Z6

# Dibujamos la red resultante:
    
d = Drawing(unit=3)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                               port_name = 'In', 
                               voltage_lbl = ('+', '$V_1$', '-'), 
                               current_lbl = '$I_1$')

d = tc2.dibujar_elemento_serie(d, Resistor, R3)

d = tc2.dibujar_tanque_RC_serie(d, R2, C2)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, R1)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C1)

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Y_{22}',  
                                          YY, 
                                          hacia_entrada = True,
                                          k_gap_width = 0.5)

d = tc2.dibujar_puerto_salida(d, 
                              port_name = 'Out', 
                              current_lbl = '$I_2$' )

display(d)
```
