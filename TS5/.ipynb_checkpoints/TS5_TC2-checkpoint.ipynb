#### Grafico 2.a) Notch Pasa Bajo

```python
import scipy.signal as sig
import numpy as np
import matplotlib.pyplot as plt

num = [1/4,0, 1]
den = [1, np.sqrt(2), 1]

sys = sig.TransferFunction(num, den) #calculado

# vector de frencuencia logaritmado
f = np.logspace(-2, 2,10000)

# obtengo la magnitud y fase de las transferencias
w, mag, phase = sig.bode(sys,f)

fig, (ax1) = plt.subplots(nrows=1 , ncols=1, sharex=True)
fig.set_size_inches(16,12)
#MODULO
ax1.semilogx(w,mag,label='Notch LP',linewidth=3.0,color='orange')
ax1.grid(True)
ax1.set_title('Magnitude Response')
ax1.set_ylabel('Magnitude[dB]')
ax1.set_ylim([-40,1])
ax1.legend()
```

#### Grafico 2.b) Elimina Banda

```python
num = [1,-0.5*np.sqrt(2), 1]
den = [1, np.sqrt(2), 1]

sys = sig.TransferFunction(num, den) #calculado

# vector de frencuencia logaritmado
f = np.logspace(-2, 2,10000)

# obtengo la magnitud y fase de las transferencias
w, mag, phase = sig.bode(sys,f)

fig, (ax1) = plt.subplots(nrows=1 , ncols=1, sharex=True)
fig.set_size_inches(16,12)
#MODULO
ax1.semilogx(w,mag,label='Elimina banda',linewidth=3.0,color='orange')
ax1.grid(True)
ax1.set_title('Magnitude Response')
ax1.set_ylabel('Magnitude[dB]')
ax1.set_ylim([-10,1])
ax1.legend()
```

#### Grafico 2.c) Pasatodo

```python
num = [1,-0.5*np.sqrt(2), 0.25]
den = [1, 0.5*np.sqrt(2), 0.25]

sys = sig.TransferFunction(num, den) #calculado

# vector de frencuencia logaritmado
f = np.logspace(-2, 2,10000)

# obtengo la magnitud y fase de las transferencias
w, mag, phase = sig.bode(sys,f)

fig, (ax2) = plt.subplots(nrows=1 , ncols=1, sharex=True)
fig.set_size_inches(16,12)

#Fase
ax2.semilogx(w, phase,label='Pasa Todo Desplazado',linewidth=3.0,color='orange')
ax2.grid(True)
ax2.set_title('Phase Response')
ax2.set_ylabel('Phase[deg]')
ax2.legend()
ax2.set_yticks([-360,-270,-180,-90,0])
ax2.set_yticklabels(['$-2\pi$','$\dfrac{-3}{2}\pi$','$\pi$','$\dfrac{-1}{2}\pi$','0'])
ax2.set_xticks([0.01,0.1,0.5,1,10,100])
ax2.set_xticklabels(['$10^{-2}$','$10^{-1}$','$\omega_{0} = 0.5$','$10^{0}$','$10^{1}$','$10^{2}$'])
```

#### Grafico 2.d) Notch Desplazado (Numerador)

```python
num = [1,0 , 4**2]
den = [1, np.sqrt(2), 1]

sys = sig.TransferFunction(num, den) #calculado

# vector de frencuencia logaritmado
f = np.logspace(-2, 2,10000)

# obtengo la magnitud y fase de las transferencias
w, mag, phase = sig.bode(sys,f)

fig, (ax2) = plt.subplots(nrows=1 , ncols=1, sharex=True)
fig.set_size_inches(16,12)

#Fase
ax2.semilogx(w, phase,label='Notch Desplazado',linewidth=3.0,color='orange')
ax2.grid(True)
ax2.set_title('Phase Response')
ax2.set_ylabel('Phase[deg]')
ax2.legend()
ax2.set_yticks([-180,-135,-90,-45,0])
ax2.set_yticklabels(['$-pi$','$\dfrac{-3}{4}\pi$','$\dfrac{-\pi}{2}$','$\dfrac{-\pi}{4}$','0'])
ax2.set_xticks([0.01,0.1,1,4,10,100])
ax2.set_xticklabels(['$10^{-2}$','$10^{-1}$','$10^{0}$','$\omega_{0} = 0.5$','$10^{1}$','$10^{2}$'])

ax2.set_ylim([-180,30])
```
