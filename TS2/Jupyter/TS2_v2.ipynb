## Simulación Circuital

### Circuito Pasabajo con Q=3, ganancia = 20dB en la banda de paso y $\omega = 1000$, es decir $f \approx 159,155 Hz$

**LTpice**
<img src="Circuito_Q3.png">

**Análisis en frecuencia**
<img src="Q3_Graph.png">

**Python**

```python
from scipy import signal
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


R1 = 1e3
R2 = 30e3
R3 = 10e3
C  = 100e-9

mpl.rcParams['figure.figsize'] = (15,15)
fig_font_size = 16
plt.rcParams.update({'font.size':fig_font_size})

num = np.array([0.0001,0.0001,-1/(R1*R3*C**2)])
den = np.array([1,1/(R2*C),1/((R3**2)*(C**2))])
# puse 0.0001 porque se enoja python si le pones  el 0 directamente


sys = signal.TransferFunction(num, den)
w, mag, phase = signal.bode(sys)

fig, (ax1,ax2) = plt.subplots(nrows=2 , ncols=1, sharex=True)

ax1.semilogx(w,mag,label='|T(w)|')
ax1.grid(True)
ax1.set_title('Magnitude Response')
ax1.set_ylabel('Magnitude[dB]')
ax1.legend()

ax2.semilogx(w, phase,color='red',label='ang[T(w)]')
ax2.grid(True)
ax2.set_title('Phase Response')
ax2.set_ylabel('Phase[deg]')
ax2.legend()

plt.xlabel('Angular frequency [rad/sec]')
```
